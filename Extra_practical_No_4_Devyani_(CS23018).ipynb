```python
clean_data=df.copy()
clean_data['high_humidity_label']=(clean_data['relative_humidity_3pm']>24.99)*1
print(clean_data['high_humidity_label'])
```

```python
##Split dataset into independent and dependant features.
X=clean_data.iloc[:,:-1]
print(X)
```

```python
y=clean_data.iloc[:,-1]
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, test_size = 0.3, random_state = 42)
```

```python
X_test
```

```python
X_train
```

```python
y_test
```

```python
y_train
```

```python
#Fit the model on the training set using the Decision Tree Classifier
humidity_classifier = DecisionTreeClassifier(max_leaf_nodes=10, random_state=0)
humidity_classifier.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(humidity_classifier, filled=True)
```

```python
#prediction
y_pred = humidity_classifier.predict(X_test)
print(y_pred)
```

```python
#Accuracy Score, Classification Report and Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
print(confusion_matrix(y_pred, y_test))
```

```python
accuracy = accuracy_score(y_pred, y_test)
```

```python
print(accuracy)
```

```python
print(classification_report(y_pred, y_test))
```
